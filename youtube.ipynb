```python
from youtube_search import YoutubeSearch    
import pandas as pd
import wget
import os
```

```python
violence =['Violence','Riots','Trafficking','Guns','Killed','Brutality','Rape','Blood','Trashed']
Adult = ['Kiss','sex','nude','devar bhabhi','MMS leaks','Porn','Romance','Fuck','bikini','beach party']
```

```python
data,label= [],[]
```

```python
for query in violence:
    results = YoutubeSearch(query, max_results=20)  
    for i in range(20):
        try :
            meta = results.videos[i]['title']
            data.append(meta)
            label.append(0)
        except IndexError:
            continue 
```

```python
file = open('violence.txt', 'w',errors='ignore')
for line in data:
    file.write(line)
file.close()
```

```python
class_data = pd.DataFrame()
```

```python
class_data['data']  = data
```

```python
class_data['label'] = label
```
